```python
import pandas as pd
```

```python
# https://pandas.pydata.org/docs/user_guide/merging.html#joining-logic-of-the-resulting-axis
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
```

```python
# result = pd.concat([df1, df2], axis=1)
result = pd.merge(df1, df2, how="cross")
```

```python
data_01 = pd.read_csv('./generated_data_01.csv')
data_02 = pd.read_csv('./generated_data_02.csv')
```

```python
data_all = pd.merge(data_01, data_02, how="cross")
```

```python
data_all.head(10)
```

```python
# https://www.geeksforgeeks.org/apply-function-to-every-row-in-a-pandas-dataframe/
from thefuzz import fuzz
def fuzzy_similarity(row):
    return fuzz.ratio(row["description_x"], row["description_y"])
#
data_all['similarity_ratio'] = data_all.apply(fuzzy_similarity, axis=1)
```

```python
data_q = data_all.query('similarity_ratio > 80')
```
